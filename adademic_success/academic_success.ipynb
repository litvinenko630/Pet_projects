### Overview

The analysis involves examining various factors to understand their influence on predicting the outcomes for students based on their demographic and academic data. This process is crucial for educational institutions, data analysts, and researchers to make informed decisions about student support, resource allocation, and educational strategies.

### Dataset Description

**Data fields:**

- **Marital Status:** Marital status of the student (1 - single, 2 - married, etc.).
- **Application Mode:** Mode of application with various codes representing different contingents and ordinances.
- **Application Order:** The order of the application, ranging from 0 (first choice) to 9 (last choice).
- **Course:** Code representing the specific course a student is enrolled in (e.g., 33 - Biofuel Production Technologies).
- **Daytime/Evening Attendance:** Attendance mode (1 - daytime, 0 - evening).
- **Previous Qualification:** Education level of the student's previous qualification (e.g., 1 - Secondary education).
- **Previous Qualification (Grade):** Grade of the previous qualification on a scale of 0 to 200.
- **Nationality:** Nationality of the student, coded numerically (e.g., 1 - Portuguese).
- **Mother's Qualification:** Education level of the mother.
- **Father's Qualification:** Education level of the father.
- **Mother's Occupation:** Occupation of the mother, coded numerically.
- **Father's Occupation:** Occupation of the father, coded numerically.
- **Admission Grade:** Admission grade of the student on a scale of 0 to 200.
- **Displaced:** Indicator if the student is displaced (1 - yes, 0 - no).
- **Educational Special Needs:** Indicator if the student has special educational needs (1 - yes, 0 - no).
- **Debtor:** Indicator if the student is a debtor (1 - yes, 0 - no).
- **Tuition Fees Up to Date:** Indicator if the student's tuition fees are up to date (1 - yes, 0 - no).
- **Gender:** Gender of the student (1 - male, 0 - female).
- **Scholarship Holder:** Indicator if the student is a scholarship holder (1 - yes, 0 - no).
- **Age at Enrollment:** Age of the student at the time of enrollment.
- **International:** Indicator if the student is an international student (1 - yes, 0 - no).
- **Curricular Units 1st Sem (Credited):** Number of curricular units credited in the 1st semester.
- **Curricular Units 1st Sem (Enrolled):** Number of curricular units enrolled in the 1st semester.
- **Curricular Units 1st Sem (Evaluations):** Number of evaluations to curricular units in the 1st semester.
- **Curricular Units 1st Sem (Approved):** Number of curricular units approved in the 1st semester.
- **Curricular Units 1st Sem (Grade):** Grade average in the 1st semester (0 to 20).
- **Curricular Units 1st Sem (Without Evaluations):** Number of curricular units without evaluations in the 1st semester.
- **Curricular Units 2nd Sem (Credited):** Number of curricular units credited in the 2nd semester.
- **Curricular Units 2nd Sem (Enrolled):** Number of curricular units enrolled in the 2nd semester.
- **Curricular Units 2nd Sem (Evaluations):** Number of evaluations to curricular units in the 2nd semester.
- **Curricular Units 2nd Sem (Approved):** Number of curricular units approved in the 2nd semester.
- **Curricular Units 2nd Sem (Grade):** Grade average in the 2nd semester (0 to 20).
- **Curricular Units 2nd Sem (Without Evaluations):** Number of curricular units without evaluations in the 2nd semester.
- **Unemployment Rate:** Unemployment rate (%).
- **Inflation Rate:** Inflation rate (%).
- **GDP:** Gross Domestic Product.
- **Target:** The outcome category at the end of the course duration (dropout, enrolled, graduate).

````python
# Path to the CSV data file
TRAIN_DATAPATH = './data/train.csv'

# Path to the Tset CSV data file for submission
TEST_DATAPATH = './data/test.csv'

# Read the training data into a pandas DataFrame
train = pd.read_csv(TRAIN_DATAPATH)

# Read the test data into a pandas DataFrame
test = pd.read_csv(TEST_DATAPATH)
````

````python
# Concatenating train and test dataframes along axis 0 (rows) to combine them into a single dataframe.
df = pd.concat([train, test], axis=0)


# Displaying the first few rows of the DataFrame
df.head()
````

````python
# Initialize empty lists to store object and non-object columns
obj = []
ints = []

# Loop through DataFrame columns
for col in df.columns:
    # Check if column data type is object
    if df[col].dtype == 'object':
        # If object, append column name, unique values count, and count of missing values to 'obj' list
        obj.append((col, df[col].nunique(), df[col].isna().sum()))
    else:
        # If non-object, append column name, unique values count, and count of missing values to 'ints' list
        ints.append((col, df[col].nunique(), df[col].isna().sum(), df[col].skew()))

# Determine the maximum length of 'obj' and 'ints' lists
max_len = max(len(obj), len(ints))

# Extend 'obj' and 'ints' lists with empty tuples to match the maximum length
obj.extend([('', '', '')] * (max_len - len(obj)))
ints.extend([('', '', '', '')] * (max_len - len(ints)))

# Create a dictionary with keys representing column categories and values representing lists of corresponding data
data = {
    'Categorical_columns': [x[0] for x in obj],
    'cat_cols_uniques': [x[1] for x in obj],
    'cat_cols_missing': [x[2] for x in obj],
    'Numeric_columns': [x[0] for x in ints],
    'int_cols_uniques': [x[1] for x in ints],
    'int_cols_missing': [x[2] for x in ints],
    'int_cols_skew': [x[3] for x in ints]
}

# Convert the dictionary into a pandas DataFrame
pd.DataFrame(data)
````

### Preprocessing the Column Names

````python
# Initialize an empty list to store the processed column names
col_list = []

# Iterate over each column name in the DataFrame
for i, j in enumerate(df.columns.tolist()):
    # Initialize an empty string for the new column name
    s = ''  
    # Iterate over each character in the column name
    for char in j:
        # Check if the character is not one of the specified characters
        if char not in "[]()'/":
            s += char  # Add the character to the new column name
        elif char == "/":
            s += ' '  # Replace '/' with a space
        else:
            s += ''  # Remove the character (no need for this else statement as it does nothing)
    # Convert the column name to lowercase and split by spaces
    s = s.lower().split()
    # Join the split parts with underscores and add to col_list
    col_list.append('_'.join(s))

# Set the DataFrame's columns to the processed column names
df.columns = col_list
````

**Custom Transformer: NumberSelector**

The `NumberSelector` class remains a custom transformer used for feature selection within the pipeline. It integrates seamlessly into scikit-learn's pipeline framework, inheriting from `BaseEstimator` and `TransformerMixin`. Here's a breakdown:

- **Initialization (`__init__`):** It takes a key parameter specifying the column to select.
- **Fit Method (`fit`):** Although required, it doesn't perform any operation, returning self to maintain pipeline compatibility.
- **Transform Method (`transform`):** It selects and returns the specified column from the input DataFrame.

This class ensures targeted column selection, enabling focused processing within the pipeline.

**Parameter Distribution for HalvingRandomSearchCV**

The dictionary `param_distributions` now defines hyperparameters only for the number of estimators for the `LGBMClassifier` to optimize through `HalvingRandomSearchCV`.

**Stratified K-Fold Cross-Validation Strategy**

The code implements Stratified K-Fold Cross-Validation with 5 folds (`StratifiedKFold`) for robust model evaluation. It splits the data into training and testing sets iteratively, ensuring a balanced class distribution in each fold.

**Feature Engineering Pipeline**

For each column in the training data, a pipeline is constructed comprising:

- **Number Selection (NumberSelector):** It selects the column.
- **Scaling (RobustScaler):** It scales the selected column.

These pipelines are aggregated using `FeatureUnion`, ensuring independent processing of each feature before merging into a single dataset.

**Model Pipeline and Hyperparameter Tuning**

The final model pipeline encompasses:

- **Feature Processing (`FeatureUnion`):** It combines feature transformation pipelines.
- **Classifier (`LGBMClassifier`):** It is a gradient boosting classifier optimized for multiclass classification with the `multi_logloss` metric.

Hyperparameter tuning employs `HalvingRandomSearchCV`, searching only through specified parameter distributions for the optimal number of estimators. This strategy efficiently explores the hyperparameter space, enhancing model performance.

**Model Evaluation**

Following model fitting, accuracy values are computed for each fold on the test set to evaluate performance. These metrics offer insights into the model's discriminatory ability and overall quality of predictions.

**Hyperparameter Optimization with Optuna**

The code further includes an Optuna optimization for hyperparameter tuning:

- **Objective Function:** It defines the hyperparameters to optimize and fits the `LGBMClassifier` with the suggested parameters.
- **Study Creation:** An Optuna study is created to maximize the objective function.
- **Optimization:** The study is optimized over a specified number of trials (epochs).
- **Best Model:** The best parameters found during optimization are used to create the final model, which is then evaluated for accuracy.

**Summary**

This code exemplifies a systematic approach to model development, incorporating feature transformation, hyperparameter optimization with `HalvingRandomSearchCV` and Optuna, and rigorous evaluation methodologies. Leveraging scikit-learn's pipeline infrastructure, cross-validation techniques, and Optuna's optimization capabilities fosters efficient model building, ultimately leading to enhanced predictive accuracy.


````python

# Enable tqdm for pandas 
tqdm.pandas()

# Define a custom transformer class called NumberSelector
class NumberSelector(BaseEstimator, TransformerMixin):
    # Initialize the transformer with a key
    def __init__(self, key):
        # Store the provided key as an instance variable
        self.key = key

    # The fit method is required but doesn't need to do anything for this transformer
    def fit(self, X, y=None):
        # Return self to allow for method chaining
        return self

    # The transform method selects and returns the column specified by the key
    def transform(self, X):
        # Use the key to select the column from the DataFrame X and return it
        return X[[self.key]]


# Define a parameter distribution for HalvingRandomSearchCV
param_distributions = {
    'classifier__n_estimators': [100, 150, 200],
}


# Initialize a Stratified K-Fold Cross-Validation with 5 folds
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# List to store AUC values for each fold
metrics = []

# Perform Stratified K-Fold Cross-Validation
for train_index, test_index in tqdm(skf.split(X, y), total=skf.get_n_splits(), desc='K-Fold Progress'):
    # Split data into training and testing sets based on the indices
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # List to store transformers for each column
    final_transformers = []
    
    # Create a pipeline for each column in the training set
    for col in X_train.columns:
        pipeline = Pipeline([
            ('selector', NumberSelector(key=col)),  # Select the column
            ('scaler', RobustScaler())  # Scale the selected column
        ])
        final_transformers.append((col, pipeline))
    
    # Combine all column transformers into a FeatureUnion
    feats = FeatureUnion(final_transformers)
    
    # Create a pipeline for feature processing
    feature_processing = Pipeline([('feats', feats)])
    
    # Create the final pipeline including the feature processing and the classifier
    pipeline = Pipeline([
        ('features', feature_processing),
        ('classifier', LGBMClassifier(objective='multiclass', num_class=3, metric='multi_logloss',
                                      random_state=42, learning_rate=0.1, 
                                      # feature_fraction=0.8, bagging_fraction=0.8, bagging_freq=5, num_leaves=31,
                                      verbosity=-1))
    ])
    
    # Create an instance of HalvingRandomSearchCV
    hyperband_search = HalvingRandomSearchCV(
        estimator=pipeline,
        param_distributions=param_distributions,
        cv=5,  # 5-fold cross-validation
        factor=3,  # Halving factor
        resource='n_samples',
        max_resources='auto',
        random_state=42
    )
    
    # Fit the HalvingRandomSearchCV on the entire dataset
    hyperband_search.fit(X_train, y_train)

    # Predict the probabilities for the test set
    y_pred = hyperband_search.predict(X_test)
    
    # Calculate the accuracy for the validation set
    accuracy = accuracy_score(y_test, y_pred)
    
    # Append the AUC value to the list
    metrics.append(accuracy)

# Print the AUC values for each fold and the mean AUC
print(f"Accuracy: {metrics}")
print(f"Mean Accuracy: {np.mean(metrics)}")
print("----------------")
````

# <a id='eighth_01'></a> Model Tuning

````python
# Create an instance of StandardScaler
scaler = RobustScaler()

# Scale the features in postlasso_df using aStandardScaler
X_scaled = scaler.fit_transform(X)

# Splitting the data into train & test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)
````

````python
# Defining the number of epochs for optimization.
epoch = 10

# Defining the objective function for optimization using Optuna.
def objective(trial):    
    # Defining hyperparameters to optimize.
    param={ 'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.1),    
            'n_estimators': trial.suggest_int('n_estimators', 50, 2500),
            'reg_alpha': trial.suggest_loguniform('reg_alpha', 0.01, 1.0),      
            'reg_lambda': trial.suggest_loguniform('reg_lambda', 0.1, 1.0), 
            # 'max_depth': trial.suggest_int('max_depth', 5, 25),
            'colsample_bytree': trial.suggest_loguniform('colsample_bytree', 0.1, 1.0)
          }
    
    # # Initializing the LightGBM model with suggested hyperparameters.
    model = LGBMClassifier(random_state=42, objective='multiclass', num_class=3, metric='multi_logloss', verbosity=-1, **param)

    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
        
    # Calculate the accuracy for the validation set
    accuracy = accuracy_score(y_test, y_pred)
    
    # Calculating and returning accuracy score.
    return accuracy

# Creating an Optuna study to maximize the objective function.
study = optuna.create_study(direction='maximize')

# Optimizing the objective function with a specified number of trials.
study.optimize(objective, n_trials=epoch)

# Printing the best parameters found during optimization.
print("Best params:", study.best_params)

# Printing the best accuracy score found during optimization.
print('Best score:', study.best_value)

# Creating the best model using the best parameters found during optimization.
best_model = LGBMClassifier(random_state=42, verbosity=-1, **study.best_params)
best_model.fit(X_train, y_train)

# Making predictions using the best model.
y_pred = best_model.predict(X_test)
        
# Calculating accuracy using the best model.
accuracy = accuracy_score(y_test, y_pred)

# Printing the accuracy achieved by the best model.
print("Accuracy:", accuracy)
````

I noticed that scaling doesn't affect model's generalyzing ability, since the ```LGBMClassifier``` is a tree-based model, specifically a gradient boosting, and tree-based models are not affected by the scale of features, unlike some other algorithms such as ```SVM``` or ```k-NN```.

Tree-based models make decisions based on thresholding individual features and comparing them with data to split into different branches of the trees. Since the splits are based on the values of individual features, not their scale, feature scaling usually does not affect the performance of tree-based models like ```LGBMClassifier```. So i chose not to scale the X.

````python
# Transform class labels for multi-class classification
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])  # Replace with your classes
n_classes = y_test_bin.shape[1]

# Probability predictions for each class
y_probs = best_model.predict_proba(X_test)

# Array for binary predictions
binary_preds = np.zeros_like(y_test_bin)

# Process each class separately
for i in range(n_classes):
    precision, recall, thresholds = precision_recall_curve(y_test_bin[:, i], y_probs[:, i])
    f_score = (2 * precision * recall) / (precision + recall)
    ix = np.argmax(f_score)
    binary_preds[:, i] = np.where(y_probs[:, i] > thresholds[ix], 1, 0)

# Convert binary predictions back to classes
predicted_classes = np.argmax(binary_preds, axis=1)

# Generate classification report
report = classification_report(y_test, predicted_classes)

# Calculate the accuracy for the validation set
print(f'Accuracy: {accuracy_score(y_test, predicted_classes)}')

# Print the classification report
print(report)
````
